```python
values=data.values()
```

```python
values
```

```python
measure=list(values)
```

```python
measure
```

```python
plt.plot(company_names,measure,marker='s')
plt.xlabel('company_name' )
plt.ylabel('measure')
plt.xticks(rotation=90)
plt.show()
```

```python
mean=sum(measure)/len(measure)
mean
```

```python
sum_Numerator=0
for i in measure:
    Numerator=(i-mean)**2
    sum_Numerator +=Numerator
Variance=sum_Numerator/(len(measure))
Variance
```

```python
Standard_deviation=np.sqrt(Variance)
Standard_deviation
```
